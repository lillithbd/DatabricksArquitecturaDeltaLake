```python
#Ejercicio 1
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

def cuadrado(num):
    return num * num

cuadrado_udf = udf(cuadrado, DoubleType())
```

```python
#Ejercicio 2
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType

def longitud(texto):
    return len(texto)

longitud_udf = udf(longitud, IntegerType())
```

```python
#Ejercicio 3
from pyspark.sql.functions import udf
from pyspark.sql.types import BooleanType

def es_par(num):
    return num % 2 == 0

es_par_udf = udf(es_par, BooleanType())
```

```python
#Ejercicio 4
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def revertir(texto):
    return texto[::-1]

revertir_udf = udf(revertir, StringType())
```

```python
#Ejercicio 5
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

def sumar_elementos(lista):
    return sum(lista)

sumar_elementos_udf = udf(sumar_elementos, DoubleType())
```

```python
#Ejercicio 6
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType
import math

def calcular_area_circulo(radio):
    return math.pi * radio**2

calcular_area_circulo_udf = udf(calcular_area_circulo, DoubleType())
```

```python
#Ejercicio 7
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def convertir_mayusculas(texto):
    return texto.upper()

convertir_mayusculas_udf = udf(convertir_mayusculas, StringType())
```

```python
#Ejercicio 8
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType

def calcular_edad(anio_nacimiento):
    return 2023 - anio_nacimiento

calcular_edad_udf = udf(calcular_edad, IntegerType())
```

```python
#Ejercicio 9
from pyspark.sql.functions import udf
from pyspark.sql.types import BooleanType

def es_vocal(letra):
    vocales = "aeiouAEIOU"
    return letra in vocales

es_vocal_udf = udf(es_vocal, BooleanType())
```

```python
#Aplicacion de ejemplo
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Inicializar SparkSession
spark = SparkSession.builder.appName("UDF Ejercicios").getOrCreate()

# Crear DataFrame de ejemplo
data = [("Alice", 25), ("Bob", 30), ("Charlie", 22)]
columns = ["Nombre", "Edad"]
df = spark.createDataFrame(data, columns)

# Aplicar UDFs a las columnas del DataFrame
df = df.withColumn("Edad al cuadrado", cuadrado_udf(col("Edad")))
df = df.withColumn("Longitud del nombre", longitud_udf(col("Nombre")))
df = df.withColumn("Es par", es_par_udf(col("Edad")))
df = df.withColumn("Nombre revertido", revertir_udf(col("Nombre")))
df = df.withColumn("Suma de edades", sumar_elementos_udf(col("Edad")))
df = df.withColumn("Área del círculo (radio=Edad)", calcular_area_circulo_udf(col("Edad")))
df = df.withColumn("Nombre en mayúsculas", convertir_mayusculas_udf(col("Nombre")))
df = df.withColumn("Edad calculada", calcular_edad_udf(col("Edad")))
df = df.withColumn("Es vocal", es_vocal_udf(col("Nombre").substr(1, 1)))

# Mostrar el DataFrame resultante
df.show()
```
